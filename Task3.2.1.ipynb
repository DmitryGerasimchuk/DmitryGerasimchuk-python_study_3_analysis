```python
import pandas as pd
```

```python
df_laptop = pd.read_csv('laptop_price.csv', encoding='cp1251')
```

```python
df_laptop.info()
```

```python
df_laptop['Cpu'].unique()
```

```python
str1 = 'Intel Core i5 2.3GHz'

str1.split()[0]
```

```python
def split_str(list_str):
    a = list_str.split()[0]
    return a

split_str(str1)
```

```python
df_laptop['Cpu_Company'] = df_laptop['Cpu'].apply(split_str)

df_laptop
```

```python
df_laptop[['Cpu', 'Cpu_Company']]
```

```python
df_laptop.head(3)
```

```python
df_laptop['Cpu_Company'].value_counts()
```

```python
df_laptop['Cpu_Company'].unique()
```

```python
df_laptop['Memory'].value_counts()
```

```python
def sum_memory(list_str):
    b = int(float(list_str.split()[0][0:-2]))
    return b

str2 = '256GB SSD +  1.0TB Hybrid '
str3 = '180GB SSD '

sum_memory(str2)
```

```python
df_laptop['Memory_Amoun'] = df_laptop['Memory'].apply(sum_memory)

df_laptop
```

```python
df_laptop['Memory'].value_counts()
```

```python
def object_memory(list_str):
    c = list_str.split()[1]
    return c

str2 = '256GB SSD +  1.0TB Hybrid'
str3 = '180GB SSD'
str4 = '512GB Flash Storage'

object_memory(str4)
```

```python
df_laptop['Memory_Type'] = df_laptop['Memory'].apply(object_memory)

df_laptop.head(5)
```

1.4 Удалите признаки Memory и ScreenResolution

```python
df_laptop.drop(columns=['Memory', 'ScreenResolution'], inplace=True)

df_laptop.info()
```
