```python
def age_to_days(x):
    return x * 365

df['Age (days_version_3)'] = df['Age'].apply(age_to_days)

df.head(10)
```

```python
import time
from tqdm import tqdm
tqdm.pandas()

def age_to_days(x):
    time.sleep(0.001)
    return x * 365

df['Age'].progress_apply(age_to_days)
```
