```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_laptop = pd.read_csv('laptop_price_for_seminar_4.csv')

df_laptop.head(5)
```

```python
plt.figure()
plt.hist(df_laptop['Memory_Amount'], bins=24)
plt.xlabel('Объем памяти')
plt.ylabel('Количество ноутбуков с заданным объемом памяти')
plt.title('Гистограмма распределения памяти')
# plt.grid()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_laptop['Price_euros'], bins=50)
plt.title('Гистограмма распределения стоимости ноутбуков')
plt.xlabel('Стоимость ноутбуков')
plt.ylabel('Количество ноутбуков с заданной стоиомстью')

# plt.grid()
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot(df_laptop['Price_euros'])
plt.title('Распределение стоимости ноутбуков')
plt.ylabel('Количество ноутбуков с заданной стоиомстью')

# plt.grid()
```

```python
plt.figure(figsize=(12, 4))
plt.hist(df_laptop['Weight'], bins=60)
plt.title('Гистограмма распределения веса ноутбуков')
plt.xlabel('Вес ноутбуков')
plt.ylabel('Количество ноутбуков с заданным весом')

# plt.grid()
```

### Задача № 3

Изучите взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров, используя сложенную или многорядовую столбчатую диаграмму

Процессоры от Samsung не изучайте

1 Постройте график в абсолютных величинах

```python
df_laptop.head()
```

```python
data_company_cpu = (pd.crosstab(df_laptop['Company'], df_laptop['Cpu_Company'])).drop(columns='Samsung').reset_index()

data_company_cpu
```

```python
plt.figure()

n_ticks = np.arange(len(data_company_cpu['Intel']))

offset = 0.2

w = 0.4

plt.bar(n_ticks - offset, data_company_cpu['Intel'], width=w)
plt.bar(n_ticks + offset, data_company_cpu['AMD'], width=w)

plt.title('Взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров')
plt.xlabel('Ноутбуки')
plt.ylabel('Количество')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data_company_cpu['Company'], rotation = 90);
```

### Задача № 4

Постройте матрицу корреляций для таблицы

```python
corr_matrix = df_laptop.corr(numeric_only=True)

corr_matrix = np.round(corr_matrix, 2)

corr_matrix = np.abs(corr_matrix)

corr_matrix
```

```python
plt.figure()
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm')
```
