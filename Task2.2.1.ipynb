```python
notebook_7700HQ = df[df['Cpu'] == 'Intel Core i7 7700HQ 2.8GHz']['Price_euros'].mean()

notebook_7700HQ
```

```python
notebook_7600U = df[df['Cpu'] == 'Intel Core i7 7600U 2.8GHz']['Price_euros'].mean()

notebook_7600U
```

```python
if notebook_7700HQ < notebook_7600U:
    print('В среднем дешевле модель ноутбука с Intel Core i7 7700HQ 2.8GHz')
else:
    print('В среднем дешевле модель ноутбука с Intel Core i7 7600U 2.8GHz')
```
