2 Выберите дом клиенту.
Клиент хочет дом:
- либо с очень красивым видом(параметр 'view' - больше) из окна, либо с видом на набережную(waterfront)
- и в очень хорошем состоянии(condition) и год постройки не меньше 1980 года.
- В какой ценовом диапазоне будут дома?

3 Выберите дом клиенту. Клиент хочет дом
- без подвала
- с двумя этажами
- стоимостью до 150000.
- Какая оценка по состоянию у таких домов в среднем?

```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df['waterfront'].unique()
```

```python
count_house_3bathrooms = df[(df['waterfront'] == 1) & (df['bathrooms'] > 3) & (df['sqft_basement'] > 0)]['id'].count()

count_house_3bathrooms
```

```python
df['sqft_basement'].unique()
```

```python
print(f'Под данный запрос соответсвует {count_house_3bathrooms} домов')
```

```python
df[(df['waterfront'] == 1) & (df['bathrooms'] > 3) & (df['sqft_basement'] > 0)]
```

```python
df['view'].unique()
```

```python
df[((df['view'] > 3) | (df['waterfront'] == 1)) & (df['condition'] > 4) & (df['yr_built'] > 1980)]
```

```python
price_min = df[((df['view'] > 3) | (df['waterfront'] == 1)) & (df['condition'] > 4) & (df['yr_built'] > 1980)]['price'].min()
price_max = df[((df['view'] > 3) | (df['waterfront'] == 1)) & (df['condition'] > 4) & (df['yr_built'] > 1980)]['price'].max()
price_mean = df[((df['view'] > 3) | (df['waterfront'] == 1)) & (df['condition'] > 4) & (df['yr_built'] > 1980)]['price'].mean()

print(f'Ценовой диапазон выбранных домов укладывается от {price_min} до {price_max}')
```

```python
condition_house = df[(df['sqft_basement'] == 0) & (df['floors'] == 2.0) & (df['price'] < 150000)]['condition'].mean()

print(f'Оценка состояния у таких домов в среднем в среднем составляет {condition_house}')
```
