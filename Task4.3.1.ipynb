### Задача № 3

Работа с датасетом "house price"

3.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
# Составаим матрицу корреляции, которая покажет зависимость стоимость недвижимости от других числовых показателей 

corr_matrix = df_house.corr(numeric_only=True)

corr_matrix = np.round(corr_matrix, 2)

corr_matrix = np.abs(corr_matrix)

corr_matrix
```

```python
# Представим тепловую карту корреляции стоимости домов от прочих числовых параметров 

plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm')
plt.title('Диаграмма № 1: Тепловая карта корреляций')
```

```python
# Получим список параментов, отсортированных по уменьшению корреляции со стоимостью домов:

corr_matrix['price'].sort_values(ascending=False)
```

Первичный вывод по графику: тепловая карта явно показывает, что следующие пять характеристик недвижимости в какой-то степени влияют на стоимость недвижимости:
- sqft_living (0.70)
- grade (0.67)
- sqft_above (0.61)

- bathrooms (0.53)

```python
# Подготовка данных для постройки диаграммы рассеяния

data_sqft_living = df_house.groupby('sqft_living')['price'].sum().reset_index()
data_sqft_living.head()
```

```python
# Построение диаграммы рассеяния для определения корреляции между общей жилой площадью и стоимостью дома

plt.figure(figsize=(10, 10))
plt.scatter(df_house['sqft_living'], df_house['price'])
plt.title('Диаграмма № 2: График разброса значений (диаграмма рассеяния)')
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость дома')
plt.grid()
```

Первичный вывод по графику: чем меньше жилая площадь дома, тем меньше стоимость дома. Причем можно наблюдать, что практически все данные сосредоточены в нижнем левом углу - чаще всего покупают дома с не самой высокой площадью. 

```python
# Построение диагрыммы рассеяния № 3: зависимость между стоимостью дома и оценкой его конструкции

plt.figure(figsize=(10, 10))
sns.jointplot(x=df_house['grade'], y=df_house['price'], kind='reg')
plt.title('Диаграмма № 3: График разброса значений')
plt.xlabel('Оценка конструкции здания')
plt.ylabel('Стоимость дома')
plt.grid()
```

Первичный вывод по графику: наблюдается прямая корреляция между стоимостью дома и значением оценки его конструкции.

```python
# Построение диагрыммы рассеяния № 4: зависимость между стоимостью дома и его площадью над уровнем земли

plt.figure(figsize=(10, 10))
sns.jointplot(x=df_house['sqft_above'], y=df_house['price'], kind='kde')
plt.title('Диаграмма № 4: График разброса значений')
plt.xlabel('Площадь дома над землей')
plt.ylabel('Стоимость дома')
plt.grid();
```

Первичный вывод по графику:отмечу, что данная диаграмма не лучшим образом показывает корреляционные связи между данными (не явно и не понятно), но использовалась данная диаграмма сугубо, чтобы проверить работоспособность настоящей диаграммы. 

```python
# Построение диаграммы с усиками № 5: зависимость между стоимостью дома и количеством ванных комнат в нем

plt.figure(figsize=(16, 8))

sns.boxplot(x=df_house['bathrooms'], y=df_house['price'], whis=1.5)

plt.title('Диаграмма № 5: Зависимость между стоимостью дома и количеством в доме ванных комнат')
plt.xlabel('Количество ванных комнат в доме')
plt.ylabel('Средняя стоимость дома');
```

Первичный вывод по графику: отмечу, что данную диаграмму целесообразно делать или не нескольких значений (количество ванных комнат) или для среднего значения ванных компат, так как, в противном случае, слишком много объектов для анализа на диаграмме. 
Если анализировать именно данную структуру графика, то можно легко сказать, что наблюдается прямая зависимость стоимости от количества ванных комнат в доме, так как чем больше ванных комнат, тем выше находится основная фигура и тем шире "усики".
